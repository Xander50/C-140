```python
import pandas as pd
credits_df=pd.read_csv('tmdb_5000_credits.csv')
movies_df=pd.read_csv('tmdb_5000_movies.csv')
credits_df.head()
```

```python
movies_df.head(5)
```

```python
credits_df.rename(columns={'movie_id':'id'},inplace=True)
credits_df.head(3)
```

```python
result_df=pd.merge(movies_df,credits_df,on='id')
result_df.head(2)
```

```python
#C-139
result_df.info()
```

```python
result_df.drop(['homepage','title_x','title_y','production_companies'],axis=1,inplace=True)
result_df.info()
```

```python
result_df.dropna(inplace=True)
result_df.info()
```

```python
R=result_df['vote_average']
v=result_df['vote_count']
C=result_df['vote_average'].mean()
print(C)
```

```python
m=result_df['vote_count'].quantile(0.9)
print(m)
```

```python
result_df['weighted_rating']=(R*v+C*m)/(v+m)
result_df[['original_title','weighted_rating']].head()
```

```python
result_df.sort_values('weighted_rating',ascending=False,inplace=True)
result_df[['original_title','weighted_rating','vote_average','vote_count','popularity']].head(10)
```

```python
import plotly.express as px
bar_plot=px.bar(result_df.head(20).sort_values('weighted_rating'),x='weighted_rating',y='original_title',orientation="h")
bar_plot.show()
```

```python
import numpy as np

def get_director(crew):
  for crew_member in crew:
    if crew_member['job'] == 'Director':
      return crew_member['name']
  return np.nan
result_df['director']=result_df['crew'].apply(get_director)
result_df[['original_title','director']].head()
```
